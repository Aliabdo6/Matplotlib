# Matplotlib lifecycle

```python
# 1st method 
fig = plt.figure() # creates a figure 
ax = fig.add_subplot()  # adds some axes
plt.show()
```

```python
# 2nd method 

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x , y)
plt.show()
```

```python
# 3rd method (recommended)
fig , ax = plt.subplots()
ax.plot(x , [24,56,868,78,33,56]);
```

```python
x = [1,2,3,4]
y = [11,99,43,87]

fig , ax = plt.subplots(figsize=(10,10)) # (width , hight)
ax.plot(x,y)

ax.set(title='Simple Plot' , 
       xlabel='xlabel',
       ylabel='ylabel' )

fig.savefig('sampl-plot.png')

```
